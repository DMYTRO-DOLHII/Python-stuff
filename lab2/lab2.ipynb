```python
import pandas as pd

data = pd.read_csv('NationalNames.csv')

print("\nTask 18: Кількість років, коли дівчаток народжувалося більше, ніж хлопчиків:")
female_vs_male_births = data.groupby(["Year", "Gender"])["Count"].sum().unstack()
years_more_female = (female_vs_male_births["F"] > female_vs_male_births["M"]).sum()
print(f"Кількість років, коли дівчаток народжувалося більше, ніж хлопчиків: {years_more_female}")
```

```python
import pandas as pd

data = pd.read_csv('NationalNames.csv')

import matplotlib.pyplot as plt

print("\nTask 19: Графік загальної кількості народжень хлопчиків та дівчаток на рік:")
plt.figure(figsize=(10, 6))
plt.plot(female_vs_male_births.index, female_vs_male_births["F"], label="Дівчатка", color="red")
plt.plot(female_vs_male_births.index, female_vs_male_births["M"], label="Хлопчики", color="blue")
plt.title("Кількість народжень хлопчиків та дівчаток на рік")
plt.xlabel("Рік")
plt.ylabel("Кількість народжень")
plt.legend()
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.savefig("gender_births_per_year.png")
plt.show()
```

```python
print("\nTask 20: Кількість гендерно нейтральних імен:")
neutral_names = data.groupby("Name")["Gender"].nunique()
neutral_names_count = len(neutral_names[neutral_names == 2])
print(f"Кількість гендерно нейтральних імен: {neutral_names_count}") 
```

```python
print("\nTask 21: Скільки разів хлопчиків називали Barbara:")
boys_named_barbara = data[(data["Name"] == "Barbara") & (data["Gender"] == "M")]["Count"].sum()
print(f"Хлопчиків, названих Barbara: {boys_named_barbara}")
```

```python
print("\nTask 23: Найпопулярніші гендерно-нейтральні імена (ті, що присутні кожного року):")
gender_neutral_names = data.groupby('Name')['Gender'].nunique()
gender_neutral_names = gender_neutral_names[gender_neutral_names > 1].index.tolist()

neutral_names_data = data[data['Name'].isin(gender_neutral_names)]

years = neutral_names_data['Year'].unique()
neutral_names_per_year = neutral_names_data.groupby('Name')['Year'].nunique()
names_in_all_years = neutral_names_per_year[neutral_names_per_year == len(years)].index.tolist()

final_data = neutral_names_data[neutral_names_data['Name'].isin(names_in_all_years)]

final_data = final_data.groupby('Name')['Count'].sum().reset_index()
final_data = final_data.sort_values(by='Count', ascending=False)

print(final_data)
```

```python
print("\nTask 25: Графіки розподілу імен John та Mary по роках:")
john_count = data[data["Name"] == "John"].groupby("Year")["Count"].sum()
mary_count = data[data["Name"] == "Mary"].groupby("Year")["Count"].sum()

plt.figure(figsize=(10, 6))
plt.plot(john_count.index, john_count.values, label="John", color="green")
plt.plot(mary_count.index, mary_count.values, label="Mary", color="purple")
plt.title("Розподіл імен John та Mary по роках")
plt.xlabel("Рік")
plt.ylabel("Кількість")
plt.legend()
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.savefig("john_mary_distribution.png")
plt.show()
```
